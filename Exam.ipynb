# Простейшие квадратурные формулы

Чтобы вычислить примерное значение определённого интеграла, когда точное получить сложно, можно воспользоваться квадратурными формулами. Пусть у нас есть интеграл

$$I = \int_a^bf(x)dx.$$

Аппроксимируем исходную функцию интерполяционным многочленом:

$$f(x) = Q_m(x) + R_m(x),$$

тогда интеграл от него порождает квардатурную формулу

$$I = \int_a^bf(x)dx \approx \int_a^bQ_m(x)dx,$$

а интеграл от остаточного члена определяет погрешность формулы

$$\epsilon = \int_a^bR_m(x)dx.$$

Последовательно подставляем в квадратурную формулу полиномы различных степеней, начиная с нулевой:

$$Q_0(x) = f(x_0),$$

$$I \approx (b-a)f(x_0).$$

Популярные точки для выбора $x_0$:

$x_o = a, \qquad I \approx (b-a)f(a), \qquad$ (формула левых прямоугольников, a);

$x_o = b, \qquad I \approx (b-a)f(b), \qquad$ (формула правых прямоугольников, б);

$x_o = \frac{a+b}{2}, \qquad I \approx (b-a)f \left (\frac{a+b}{2} \right), \qquad$ (формула средних прямоугольников, в).


Если для узлов интерполирования полинома первой степени выбрать точки a и b, то квадратурная формула примет вид формулы трапеций (г):
$$I \approx \frac{b-a}2(f(b) + f(a)). $$

А если для полинома второй степени выбрать точки $x_0 = a$, $x_1 = \frac{b-a}2$ и $x_2 = b$, то при интегрировании получится квадратурная формула Симпсона:

$$ I \approx \frac{b-a}2 \left( f(a) + 4f \left (\frac{b-a}2 \right) + f(b) \right) $$

Проведём расчёт погрешностей для этих формул, используя в рассуждениях две теоремы из курса математического анализа.

__Теорема 1__ Пусть есть две непрерывные на $[a, b]$ функции $f(x)$ и $g(x)$, причём $g(x)$ знакопостоянна. Тогда найдётся точка $c \in [a, b]$ такая, что

$$ \int_a^bf(x)g(x)dx = f(c) \int_a^bg(x)dx.$$

__Теорема 2__ Пусть $f(x)$ непрерывна на $[a, b]$ и заданы $N$ точек $x_k \in [a, b]$. Тогда найдётся точка $\eta \in [a, b]$ такая, что

$$ \frac1n \sum_{k = 1}^N f(x_k) = f(\eta)$$

Для интерполяционного полинома нулевой степени остаточный член примет вид

$$ R_0(x) = \frac{x-x_0}{1!}f'(\eta), $$

причём $\eta = \eta(x)$. Подставляем значения $x_0$, пользуясь тем, что на $[a, b]$ значение $(x-a)$ знакопостоянно:

$$\epsilon_{лев} = \int_a^b (x-a)f'(\eta)dx = f'(\eta^*) \int_a^b(x-a)dx = \frac{(b-a)^2}2 f'(\eta^*);$$

$$\epsilon_{прав} = \int_a^b (x-b)f'(\eta)dx = f'(\eta^*) \int_a^b(x-b)dx = -\frac{(b-a)^2}2 f'(\eta^*);$$

$$\epsilon_{средн} = \int_a^b (x-\frac{a+b}2)f'(\eta)dx.$$

Разложим $f(x)$ в многочлен Тейлора в точке $\frac{a+b}2$:

$$f(x) \approx f \left( \frac{a+b}2 \right) + \frac{x - \frac{a+b}2}{1!}f'\left( \frac{a+b}2\right) + \frac{{\left( x - \frac{a+b}2 \right)}^2}{2!}f''(\eta^*)$$

Интеграл от первого слагаемого — это формула средних прямоугольников, интеграл от второго равен нулю, а от третьего как раз показывает погрешность. Таким образом,

$$\epsilon_{средн} = \frac12\int_a^b\left( x - \frac{a+b}2 \right)^2f''(\eta^*)dx = \frac{f''(\eta^*)}2 \int_a^b \left( x - \frac{a+b}2 \right)^2dx = \frac{(b-a)^3}{24}f''(\eta^*).$$

Теперь оценим погрешность для формулы трапеций, для этого интегрируем остаточный член полинома первой степени:

$$\epsilon_ {трап} = \int_a^b \frac{(x-a)(x-b)}{2!}f''(\eta)dx = f''(\eta^*) \int_a^b \frac{(x-a)(x-b)}{2!} = -\frac{(b-a)^3}{12}f''(\eta^*).$$

Осталось оценить погрешность для отстаточного члена интерполяционного многочлена второй степени:

$$R_2(x) = \frac{(x-a) \left( x - \frac{a+b}2 \right)(x-b)}{3!}f'''(\eta).$$

Чтобы оценить погрешность для формулы Симпсона, используются не теоремы о среднем, результат приведём без вывода:

$$\epsilon_{Симп} = -\frac{(b-a)^5}{2880}f^{(4)}(\eta)$$

В большинстве случаев подынтегральная функция не описывается удовлетворительно полиномами первой или второй степени. Поэтому для достижения необходимой точности исходный промежуток разбивается на такие малые промежутки, где указанная аппроксимация удачна, на каждом из этих промежутков применяется выбранная квадратурная формула, а результаты складываются. Такие формулы получили название составных (или больших) квадратурных формул.

# Составные квадратурные формулы

Разобьем исходный промежуток на $N$ равных промежутков $[x_k, x_{k+1}]$:

$$ h = \frac{b-a}N, \quad x_k = x_0 + kh, \quad x_0 = a, \quad x_N = b$$

На каждом таком промежутке применим формулу левых прямоугольников и результаты сложим:

$$ I_k = \int_{x_k}^{x_{k+1}} f(x)dx \approx (x_{k+1} - x_k)f(x_k) = \frac{b-a}Nf(x_k);$$

$$I_{лев. прям.} = \sum_{k=0}^{N-1}I_k \approx \frac{b-a}N \sum_{k=0}^{N-1}f(x_k) $$

Проведём аналогичные рассуждения для формул правых и средних прямоугольников, а также для формулы трапеций:

1. правые прямоугольники:

$$ I_k = \int_{x_k}^{x_{k+1}} f(x)dx \approx (x_{k+1} - x_k)f(x_{k+1}) = \frac{b-a}Nf(x_{k+1});$$

$$I_{прав. прям.} = \sum_{k=0}^{N-1}I_k \approx \frac{b-a}N \sum_{k=0}^{N-1}f(x_{k+1}) = \frac{b-a}N \sum_{k=1}^Nf(x_k) $$

2. средние прямоугольники:

$$ I_k = \int_{x_k}^{x_{k+1}} f(x)dx \approx (x_{k+1} - x_k)f\left(x_k+\frac h2\right) = \frac{b-a}N f\left(x_k+ \frac h2\right)$$

$$I_{средн. прям.} = \sum_{k=0}^{N-1}I_k \approx \frac{b-a}N \sum_{k=0}^{N-1}f\left(x_k+ \frac h2\right) = \frac{b-a}N \sum_{k=0}^{N-1}f\left(x_k+ \frac{b-a}{2N}\right);$$

3. трапеции:

$$ I_k = \int_{x_k}^{x_{k+1}} f(x)dx \approx \frac{(x_{k+1} - x_k)}2(f(x_{k+1} + f(x_k)) = \frac{b-a}{2N}(f(x_{k+1} + f(x_k))$$

$$ I_{трап.} = \sum_{k=0}^{N-1}I_k \approx \frac{b-a}{2N} \sum_{k=0}^{N-1}(f(x_{k+1} + f(x_k)) = \frac{b-a}{2N} \left(f(a) + 2\sum_{k=1}^{N-1}f(x_k) + f(b)\right).$$

Эти формулы называются составными квадратурными формулами левых, правых, средних прямоугольников и трапеций. Чтобы получить составную формулу Симпсона, будем выбирать $N$ всегда чётным, а исходный промежуток разобьём на $N/2$ равных промежутков $[x_k, x_{k+2}]$ длиной $2(b-a)/N$. На каждом из них интеграл равен

$$I_k = \int_{x_k}^{x_{k+2}} f(x)dx \approx \frac{x_{k+2} - x_k}6 \left( f(x_{k+2}) +4f(x_{k+1}) + f(x_k) \right) = \frac{b-a}{3N}\left( f(x_{k+2}) +4f(x_{k+1}) + f(x_k) \right).$$

Суммируем по всем промежуткам и получаем составную формулу Симпсона:

$$I_{Симпс.} = \frac{b-a}{3N}(f(a)+4(f_1+f_3+f_5+...+f_{N-1})+2(f_2+f_4+...+f_{N-2})+f(b))$$

## Оценка погрешности
Для оценки погрешности составных формул вычисляем погрешность каждого малого участка и результаты складываем. Для левых прямоугольников:

$$ \epsilon_k = \frac{(x_{k+1} - x_k)^2}2f'(\eta_k) = \frac{(b-a)^2}{2N^2}f'(\eta_k),$$

Суммируя и применяя вторую теорему о среднем к выражению в квадратных скобках, имеем:

$$ \epsilon_{лев. прям.} = \sum_{k=1}^N \epsilon_k = \frac{(b-a)^2}{2N} \left[ \frac1N \sum_{k=1}^Nf'(\eta_k) \right] = \frac{(b-a)^2}{2N}f'(\eta).$$

Повторяя аналогичные вычисления с формулой для правых прямоугольников получаем:

$$ \epsilon_{прав. прям.} = - \frac{(b-a)^2}{2N}f'(\eta).$$

Для средних прямоугольников:
    
$$ \epsilon_k = \frac{(x_{k+1} - x_k)^3}{24}f''(\eta_k) = \frac{(b-a)^3}{24N^3}f''(\eta^*),$$

$$ \epsilon_{средн. прям.} = \sum_{k=1}^N \epsilon_k = \frac{(b-a)^3}{24N^2} \left[ \frac1N \sum_{k=1}^Nf'(\eta_k) \right] = \frac{(b-a)^2}{24N^2}f''(\eta).$$

Для трапеций:

$$ \epsilon_k = -\frac{(x_{k+1} - x_k)^3}{12}f''(\eta_k) = \frac{(b-a)^3}{12N^3}f''(\eta^*),$$

$$ \epsilon_{трап.} = \sum_{k=1}^N \epsilon_k = - \frac{(b-a)^3}{12N^2} \left[ \frac1N \sum_{k=1}^Nf'(\eta_k) \right] = - \frac{(b-a)^2}{12N^2}f''(\eta).$$

Для оценки погрешности составной формулы Симпсона следует учитывать, что длина каждого малого участка в два раза больше, чем ранее, а число таких участков в два раза меньше — N/2

$$ \epsilon_k = -\frac {(x_{k+2} - x_k)^5}{2880} f^{(4)}(\eta_k) = -\frac{(b-a)^5}{90N^5} f^{(4)}(\eta_k),$$

$$ \epsilon_{Симпс} = \sum_{k=1}^{N/2} \epsilon_k = -\frac{(b-a)^5}{180N^5} \left[\frac1{N/2} \sum_{k=1}^{N/2} f^{(4)}(\eta_k) \right] = - \frac{(b-a)^5}{180N^5}f^{(4)}(\eta).$$
